```python
bruteforce = []
directory = 'brute_force_result'
for filename in os.listdir(directory):
    fn = os.path.join(directory, filename)
    with open(fn) as f:
        for line in f:
            pass
        last_line = line
        last = last_line.split()
        kc = int(re.findall(r'\d+', filename.split('_')[0])[0])
        mc = int(re.findall(r'\d+', filename.split('_')[1])[0])
        nc = int(re.findall(r'\d+', filename.split('_')[2])[0])
        
        bruteforce.append([[kc, mc, nc], float(last[2])])
```

```python
kc = [ [[k, m, n], g] for [[k, m, n], g] in bruteforce if (m == 1024 and n == 2048) ]

kc_1 = [ [[k, m, n], g] for [[k, m, n], g] in bruteforce if (m == 256 and n == 1024) ]

kc_2 = [ [[k, m, n], g] for [[k, m, n], g] in bruteforce if (m == 512 and n == 1024) ]
```

```python
x_kc = [ k for [[k, m, n], g] in kc]
y_kc = [ g for [[k, m, n], g] in kc]

x_kc_1 = [ k for [[k, m, n], g] in kc_1]
y_kc_1 = [ g for [[k, m, n], g] in kc_1]

x_kc_2 = [ k for [[k, m, n], g] in kc_2]
y_kc_2 = [ g for [[k, m, n], g] in kc_2]
```

```python
xpoints_kc = np.array(x_kc)
ypoints_kc = np.array(y_kc)

xpoints_kc1 = np.array(x_kc_1)
ypoints_kc1 = np.array(y_kc_1)

xpoints_kc2 = np.array(x_kc_2)
ypoints_kc2 = np.array(y_kc_2)


#f = plt.figure()
#f.set_figwidth(20)
#f.set_figheight(10)

plt.xlabel('kc')
plt.ylabel('GFLOPS')

plt.xticks(xpoints_kc1)
plt.plot(x_kc, y_kc, "b*", label = "m == 1024 and n == 2048")
plt.plot(xpoints_kc1, ypoints_kc1, "r*", label = "m == 256 and n == 1024")
plt.plot(xpoints_kc2, ypoints_kc2, "g*", label = "m == 512 and n == 1024")

plt.legend()
    
plt.show()
```

```python
xpoints_mc = np.array(x_mc)
ypoints_mc = np.array(y_mc)

xpoints_mc1 = np.array(x_mc_1)
ypoints_mc1 = np.array(y_mc_1)

xpoints_mc2 = np.array(x_mc_2)
ypoints_mc2 = np.array(y_mc_2)


f = plt.figure()
#f.set_figwidth(8)
#f.set_figheight(10)

plt.xlabel('mc')
plt.ylabel('GFLOPS')

plt.xticks(xpoints_mc1)
plt.plot(x_mc, y_mc, "b*", label = "k == 32 and n == 1024")
plt.plot(xpoints_mc1, ypoints_mc1, "r*", label = "k == 64 and n == 1024")
plt.plot(xpoints_mc2, ypoints_mc2, "g*", label = "k == 64 and n == 2048")

plt.legend()
    
plt.show()
```

```python
ours_size = []
ours_result = []
with open("benches/bench_blis.txt") as f:
    for line in f:
        ours_size.append(int(line.split()[0]))
        ours_result.append(float(line.split()[1]))

blas_size = []
blas_result = []
with open("benches/bench_blas.txt") as f:
    for line in f:
        blas_size.append(int(line.split()[0]))
        blas_result.append(float(line.split()[1]))

naive_size = []
naive_result = []
with open("benches/bench_naive.txt") as f:
    for line in f:
        naive_size.append(int(line.split()[0]))
        naive_result.append(float(line.split()[1]))
```

```python
xpoints_ours = np.array(ours_size)
ypoints_ours = np.array(ours_result)

xpoints_blas = np.array(blas_size)
ypoints_blas = np.array(blas_result)

xpoints_naive = np.array(naive_size)
ypoints_naive = np.array(naive_result)

f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)

plt.xlabel('SIZE')
plt.ylabel('GFLOPS')

plt.plot(xpoints_ours, ypoints_ours, "b-*", label = "ours")

plt.plot(xpoints_blas, ypoints_blas, "r-*", label = "blas")

plt.plot(xpoints_naive, ypoints_naive, "g-*", label = "naive")

plt.legend(prop={'size': 30})

plt.show()
```
